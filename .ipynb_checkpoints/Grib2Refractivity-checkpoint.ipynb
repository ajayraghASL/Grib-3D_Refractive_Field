```python
import numpy as np
import matplotlib.pyplot as plt
import pygrib 
from mpl_toolkits.mplot3d import Axes3D
import math
import matplotlib.colors as mcolors
```

```python
relativehumidity = pygrib.open('ncum_imdaa_reanl_HR_RH-prl_2018041912-2018041918.grb2')
# pressure = pygrib.open('/home/ajay/work/CAT/IMD data/ncum_imdaa_reanl_HR_PRES-sfc_2018041914-2018041915.grb2')
temperature = pygrib.open('ncum_imdaa_reanl_HR_TMP-prl_2018041912-2018041918.grb2')
```

```python
heat = []
humid = []
pressure = []
for i in range(24):
    grbtemp = temperature.select()[i]
    grbrh = relativehumidity.select()[i]
    pres = grbtemp.level
    humid.append(grbrh.values)
    heat.append(grbtemp.values)
    pressure.append(pres)
heat = np.array(heat)
humid = np.array(humid)
pressure = np.array(pressure)
lats, lons = grbtemp.latlons()
```

```python
print(heat.shape)
print(humid.shape)
print(lats.shape)
print(lons.shape)
print(pressure)
pressure.shape
np.savetxt('latitudes.txt',lats)
np.savetxt('longitudes.txt',lons)
```

```python
def hypsometric(pressure,heat):
    height = []
    for i in range(pressure.shape[0]):
        p = pressure[i]
        p0 = 1013.25
        t = np.mean(heat[i])
        h = ((((p0/p)**(1/5.257))-1)*t)/0.0065
        h = h/1000
        height.append(h)
    return height
```

```python
altitude = hypsometric(pressure,heat)
altitude
np.savetxt('altitude.txt',altitude)
```

```python
def EF(p,t):
    tc = t-273.15
    ef = 1+(7.2+p*(0.0320+(5.9*10**(-6))*tc**2))
    return ef
def ES(p,t):
    a = 6.1121
    b = 18.678
    c = 257.14
    d = 234.5
    tc = t-273.15
    ef = EF(p,t)
    es = ef*a*math.exp(((b-(tc/d))/(tc+c))*tc)
    return es
def E(h,p,t):
    es = ES(p,t)
    e = (h*es)/100
    return e
def N(h,p,t):
    e = E(h,p,t)
    n = 77.6*(p/t)-5.6*(e/t)+(3.75*10**5)*(e/t**2)
    return n
    
```

```python
nfield = np.zeros_like(heat)
for i in range(heat.shape[0]):
    for j in range(heat.shape[1]):
        for k in range(heat.shape[2]):
            t = heat[i,j,k]
            p = pressure[i]
            h = humid[i,j,k]
            Ni = N(h,p,t)
            n = 1+(Ni*10**(-6))
            nfield[i,j,k] = n
# print(nfield)
# for i in range(nfield.shape[0]):
#     np.savetxt('nfield%01d.txt'%i,nfield[i])
```

```python
values = np.unique(nfield)
values.shape
```
